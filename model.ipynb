**Data Cleaning & Preprocessing**

```python
df['arr_delay'].isnull().sum()
```

```python
df_clean = df.dropna(subset=['arr_delay', 'dep_delay'])
```

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.drop(columns=['tailnum'])
df_clean['air_time'] = df_clean['air_time'].fillna(df_clean['air_time'].median())
```

**Create Target Variable**

Define Delay Rule

A flight is considered delayed if:

Arrival delay > 15 minutes

```python
df_clean['Delayed'] = np.where(df_clean['arr_delay'] > 15, 1, 0)
```

**Check class distribution**

```python
df_clean['Delayed'].value_counts()
```

**Remove Leakage Columns**

```python
leakage_cols = ['arr_time', 'sched_arr_time', 'arr_delay']
df_clean = df_clean.drop(columns=leakage_cols)
```

**Select Final Feature Set**

```python
features = [
    'year', 'month', 'day',
    'sched_dep_time',
    'carrier',
    'origin', 'dest',
    'distance',
    'air_time',
    'hour'
]

X = df_clean[features]
y = df_clean['Delayed']
```

Data cleaning was performed by removing cancelled and diverted flights with missing delay values.
A binary target variable Delayed was created based on a 15-minute arrival delay threshold.
Data leakage was avoided by removing post-arrival information.

# Feature Engineering & Encoding

**Create New Time-Based Features**

```python
df_clean['date'] = pd.to_datetime(df_clean[['year','month','day']])
df_clean['is_weekend'] = (df_clean['date'].dt.weekday >= 5).astype(int)
df_clean = df_clean.drop(columns=['date'])
```

**Peak Hour Indicator**

```python
df_clean['is_peak_hour'] = ((df_clean['hour'].between(7,9)) |
                             (df_clean['hour'].between(16,19))).astype(int)

```

**Scheduled Departure Hour (Cleaner)**

```python
df_clean['sched_dep_hour'] = df_clean['sched_dep_time'] // 100
df_clean = df_clean.drop(columns=['sched_dep_time'])
```

```python
import os

# Create folder to store dropdown options
os.makedirs('options', exist_ok=True)

# Save unique carriers
df['carrier'].drop_duplicates().sort_values().to_csv(
    'options/carriers.txt',
    index=False,
    header=False
)

# Save unique origins (airports)
df['origin'].drop_duplicates().sort_values().to_csv(
    'options/origins.txt',
    index=False,
    header=False
)

# Save unique destinations (airports)
df['dest'].drop_duplicates().sort_values().to_csv(
    'options/destinations.txt',
    index=False,
    header=False
)

# Save unique airline names
df['name'].drop_duplicates().sort_values().to_csv(
    'options/airline_names.txt',
    index=False,
    header=False
)

# Save unique flight numbers
df['flight'].drop_duplicates().sort_values().to_csv(
    'options/flights.txt',
    index=False,
    header=False
)
```

**Separate Features & Target Again**

```python
X = df_clean.drop(columns=['Delayed'])
y = df_clean['Delayed']
```

```python
X = X.drop(columns=['dep_delay'], errors='ignore')
X = X.drop(columns=['time_hour'], errors='ignore')
```

**Encode Categorical Variables**

```python
cat_cols = ['carrier', 'origin', 'dest', 'name']
```

```python
from sklearn.preprocessing import LabelEncoder
carrier_encoder = LabelEncoder()
X['carrier'] = carrier_encoder.fit_transform(X['carrier'])
origin_encoder = LabelEncoder()
X['origin'] = origin_encoder.fit_transform(X['origin'])
dest_encoder = LabelEncoder()
X['dest'] = dest_encoder.fit_transform(X['dest'])
name_encoder = LabelEncoder()
X['name'] = name_encoder.fit_transform(X['name'])
```

```python
# Drop 'id' column from feature set X
X = X.drop(columns=['id'])
```

```python
X.head()
```

```python
X.duplicated().sum()
```

Balance the imbalanced data using SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_sm, y_sm = smote.fit_resample(X, y)
```

Feature engineering was performed by extracting time-based indicators such as peak hours and weekends. Categorical variables including airline and airports were encoded using label encoding.

# Split the dataset

**Trainâ€“Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_sm, y_sm, test_size=0.3, stratify=y_sm, random_state=42
)
```

```python
y_train.value_counts()
```

# Model Building (Training Machine Learning Models)

Why Multiple Models?

We train more than one model to:

Compare performance
Choose the best model
Show depth in your project

We will train:
1. Logistic Regression (baseline)
2. Random Forest (powerful, non-linear)

**Model 1: Logistic Regression (Baseline)**

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(
    max_iter=2000,
    class_weight={0: 1, 1: 3}
)

log_model.fit(X_train, y_train)
```

```python
y_pred_log = log_model.predict(X_test)
y_prob_log = log_model.predict_proba(X_test)[:, 1]
```

**Model 2: Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    min_samples_split=10,
    class_weight={0: 1, 1: 3},
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]
```

**Feature Importance**

```python
importances = pd.Series(
    rf_model.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

importances.head(10)
```

Two classification models were trained: Logistic Regression as a baseline model and Random Forest for capturing complex patterns. Class weights were applied to address data imbalance. Random Forest also provided feature importance for interpretability.

# Model Evaluation & Performance Comparison

**Why Evaluation Is Important**
Accuracy alone is not sufficient, especially with imbalanced data.
So we use multiple metrics.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)
```

**Evaluate Logistic Regression**

```python
print("Logistic Regression Evaluation")

print("Accuracy:", accuracy_score(y_test, y_pred_log))
print("Precision:", precision_score(y_test, y_pred_log))
print("Recall:", recall_score(y_test, y_pred_log))
print("F1 Score:", f1_score(y_test, y_pred_log))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_log))
```

**Classification Report**

```python
print(classification_report(y_test, y_pred_log))
```

**Confusion Matrix (Logistic Regression)**

```python
cm_log = confusion_matrix(y_test, y_pred_log)

plt.figure(figsize=(5,4))
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

**Evaluate Random Forest**

```python
print("Random Forest Evaluation")

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall:", recall_score(y_test, y_pred_rf))
print("F1 Score:", f1_score(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
```

**Classification Report**

```python
print(classification_report(y_test, y_pred_rf))
```

**Confusion Matrix (Random Forest)**

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(5,4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

**ROC Curve Comparison**

```python
fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_log)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.figure(figsize=(6,5))
plt.plot(fpr_log, tpr_log, label="Logistic Regression")
plt.plot(fpr_rf, tpr_rf, label="Random Forest")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.show()
```

**Model Comparison Table**

```python
comparison = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest"],
    "Accuracy": [
        accuracy_score(y_test, y_pred_log),
        accuracy_score(y_test, y_pred_rf)
    ],
    "Precision": [
        precision_score(y_test, y_pred_log),
        precision_score(y_test, y_pred_rf)
    ],
    "Recall": [
        recall_score(y_test, y_pred_log),
        recall_score(y_test, y_pred_rf)
    ],
    "F1 Score": [
        f1_score(y_test, y_pred_log),
        f1_score(y_test, y_pred_rf)
    ],
    "ROC-AUC": [
        roc_auc_score(y_test, y_prob_log),
        roc_auc_score(y_test, y_prob_rf)
    ]
})

comparison
```
